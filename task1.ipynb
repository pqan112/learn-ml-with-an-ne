### Data Visualization

```python
import matplotlib.pyplot as plt
```

#### Line chart

```python


# Sample data
x = [1,2,3,4,5]
y = [5,10,15,20,25]

plt.plot(x, y)

plt.xlabel('độ ngu', )
plt.ylabel('số tuổi')
plt.title('Simple Line Plot')

# hiển thị số nguyên
plt.xticks(x)
plt.yticks(y)

plt.show()
```

#### Bar chart

```python
# Số liệu trên phải là số ok?

rainfall = [3107, 9876, 9999]
city = ['Ho Chi Minh', 'Ha Noi', 'Da Nang']

plt.bar(city, rainfall)

plt.xlabel('City')
plt.ylabel('Rainfall')

plt.show()
```

#### Customize chart

```python
x=[2000,2005,2010,2015, 2020, 2025]

y1=[60,55,60,60,65, 80]
y2=[60,55,55,60,80, 80]
y3=[60,50,55,70,70, 75]

plt.title('The growth rate of three cities from 2000-2025')
plt.plot(x, y1, label="Ho Chi Minh")
plt.plot(x, y2, label="Ha Noi")
plt.plot(x, y3, label="Da Nang")
plt.legend()

plt.show()
```

#### Work with data frame

```python
import pandas as pd

# Dữ liệu tăng trưởng theo từng năm
data = {
    'Year': [2000,2005,2010,2015, 2020, 2025],
    'Ha Noi': [60,55,55,60,80, 80],
    'Da Nang': [60,50,55,70,70, 75],
    'Ho Chi Minh': [60,55,60,60,65, 80]
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

plt.plot(df['Year'], df['Ha Noi'], label='Ha Noi')
plt.plot(df['Year'], df['Da Nang'], label='Da Nang')
plt.plot(df['Year'], df['Ho Chi Minh'], label='Ho Chi Minh')

plt.title("The growth rate of three cities")
plt.xlabel("Year")
plt.ylabel("Growth rate")
plt.legend()
plt.show()
```
